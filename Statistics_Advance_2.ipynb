2. What is the P-value in hypothesis testing, and how should it be interpreted

-> In hypothesis testing, the p-value is a key statistical measure used to decide whether the observed data provides enough evidence to reject the null hypothesis (Hâ‚€).

âœ… What is the P-value?

The p-value is the probability of obtaining a test statistic (or result) at least as extreme as the one observed, assuming the null hypothesis is true.

In simple words:

ðŸ‘‰ It tells us how likely our sample result is, if the null hypothesis were actually true.

âœ… How to Interpret the P-value

âš ï¸ Important Notes About P-value

ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences between the means of three or more groups.

It answers the question:

Why not use multiple t-tests?

Using several t-tests increases Type I error (false positives).

ANOVA controls the overall error rate and provides one test for comparing multiple means.

ANOVA generalizes the two-sample hypothesis test by comparing variances between groups and within groups.

If the between-group variance is large compared to the within-group variance, it suggests that the group means are different.

5. Write a Python program to perform a one-sample Z-test and interpret the
   result for a given dataset.


-> Python Program: One-Sample Z-Test
Goal:

Test whether the sample mean is significantly different from a known population mean when population standard deviation is known.

ðŸ“Œ Python Code
import numpy as np
from math import sqrt
from scipy.stats import norm

# Sample data
data = [12, 15, 14, 10, 13, 16, 14, 15]

# Given population mean and population standard deviation
population_mean = 13
population_std = 2

# Step 1: Calculate sample mean and sample size
sample_mean = np.mean(data)
n = len(data)

# Step 2: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / sqrt(n))

# Step 3: Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Print results
print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("The sample mean is significantly different from the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("There is no significant difference between sample mean and population mean.")

-> Below is a long, detailed explanation + Python code to demonstrate the Central Limit Theorem (CLT) using samples generated from a non-normal distribution (Exponential distribution).

â­ Central Limit Theorem (CLT) Demonstration Using Python

The Central Limit Theorem states that:

When we draw many random samples from ANY population with a finite mean and variance, the distribution of the sample means approaches a normal distribution, even if the original population is not normal.

To illustrate this, we will:

Generate data from a non-normal distribution (Exponential distribution).

Draw repeated random samples of size n.

Compute the mean of each sample.

Plot:

The original non-normal distribution

The distribution of sample means (which becomes normal as sample size increases)

âœ… Python Program (with full explanation)
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data from a non-normal distribution (Exponential)
population_size = 100000
population = np.random.exponential(scale=1, size=population_size)

# Step 2: Set sample parameters
sample_size = 50      # size of each sample
num_samples = 1000    # number of samples

sample_means = []

# Step 3: Perform repeated sampling
for i in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step 4: Plot the original distribution and sample mean distribution
plt.figure(figsize=(14, 6))

# Plot 1: Original Non-Normal Distribution
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, density=True)
plt.title("Original Distribution (Exponential)\nHighly Right-Skewed")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot 2: Distribution of Sample Means (Approx. Normal)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True)
plt.title(f"Distribution of Sample Means (CLT Demonstration)\nSample Size = {sample_size}")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

9. Write a Python function to calculate and visualize the confidence interval
for a sample mean.


-> Calculates the confidence interval for a sample mean
âœ… Visualizes the mean and the interval on a graph
âœ… Works for any dataset, confidence level, and known/unknown population standard deviation

â­ Python Function: Confidence Interval Calculation + Visualization
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

def confidence_interval_plot(data, confidence=0.95, population_std=None):
    """
    Calculates and visualizes the confidence interval for the sample mean.
    
    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default = 95%)
        population_std (float): Provide only if population standard deviation is known.
    """

    data = np.array(data)
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)

    # Standard Error
    if population_std is not None:
        se = population_std / np.sqrt(n)
        z_value = norm.ppf((1 + confidence) / 2)
        margin_of_error = z_value * se
        method = "Z-Interval"
    else:
        se = sample_std / np.sqrt(n)
        t_value = t.ppf((1 + confidence) / 2, df=n-1)
        margin_of_error = t_value * se
        method = "T-Interval"

    # Confidence Interval
    lower = sample_mean - margin_of_error
    upper = sample_mean + margin_of_error

    # Print results
    print(f"Method Used: {method}")
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"{confidence*100:.0f}% Confidence Interval: ({lower:.4f}, {upper:.4f})")

    # Visualization
    plt.figure(figsize=(10, 4))
    plt.axvline(sample_mean, color='blue', label='Sample Mean')
    plt.axvline(lower, color='red', linestyle='--', label='Lower CI')
    plt.axvline(upper, color='red', linestyle='--', label='Upper CI')

    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.xlabel("Value")
    plt.legend()
    plt.grid(True)
    plt.show()

    return lower, upper

import numpy as np
from scipy.stats import chisquare

# Observed frequencies

else:
    print("\nConclusion: Reject H0.")
    print("Observed distribution is significantly different from expected.")
